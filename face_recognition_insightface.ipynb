```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import insightface
from insightface.app import FaceAnalysis
from insightface.data import get_image as ins_get_image
```

```python
def process_images(image1_path, image2_path):
  img1 = cv2.imread(image1_path)
  img2 = cv2.imread(image2_path)

  # to install model
  detector = insightface.model_zoo.get_model('model.onnx', download=True)
  app = FaceAnalysis()

  # to show images
  plt.imshow(img1[:,:,::-1])
  plt.show()
  plt.imshow(img2[:,:,::-1])
  plt.show()

  # to prepare images 
  app.prepare(ctx_id=0)
  faces_1 = app.get(img1)[0]['embedding']
  faces_2 = app.get(img2)[0]['embedding']

  # to compare image similarity
  print(f'Similarity score: {detector.compute_sim(faces_1, faces_2)}')
```

```python
image1_path = '/content/sheldon_face3.jfif'
image2_path = '/content/sheldon_face2.jfif'
process_images(image1_path, image2_path)
```
